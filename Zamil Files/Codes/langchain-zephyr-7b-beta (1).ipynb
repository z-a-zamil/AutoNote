```python
from langchain.llms import HuggingFaceHub
from langchain.prompts import PromptTemplate

# Load the model from Hugging Face Hub
model = HuggingFaceHub(repo_id="HuggingFaceH4/zephyr-7b-beta", model_kwargs={"device": "cuda"})

# Define the text to summarize
text = """
This is a long article about the latest developments in artificial intelligence.
It discusses the recent breakthroughs in natural language processing, computer vision, and reinforcement learning.
The article also explores the potential implications of these advances for various industries and society as a whole.
It provides insights from leading experts in the field and highlights the challenges and ethical considerations surrounding AI.
"""

# Create a prompt template for summarization
prompt_template = PromptTemplate(input_variables=["text"], template="Summarize the following text:\n\n{text}")

# Generate the prompt with the text to summarize
prompt = prompt_template.format(text=text)

# Generate the summary
summary = model.generate([prompt], max_length=100, min_length=30, do_sample=False, early_stopping=True)

# Print the summary
print(summary[0])
```

```python
# Zero Shot

prompt_template = """
Summarize the following text:

{text}
"""

prompt = PromptTemplate(template=prompt_template, input_variables=["text"])

text = """
This is a long article about the latest developments in artificial intelligence.
It discusses the recent breakthroughs in natural language processing, computer vision, and reinforcement learning.
The article also explores the potential implications of these advances for various industries and society as a whole.
It provides insights from leading experts in the field and highlights the challenges and ethical considerations surrounding AI.
"""

summarization_prompt = prompt.format(text=text)
summary = llm(summarization_prompt, max_length=100, min_length=30, do_sample=False)[0]["summary_text"]
print(summary)
```
