```python
from nltk.tokenize import word_tokenize
from nltk.translate.bleu_score import corpus_bleu
from rouge import Rouge
import nltk
nltk.download('punkt')

def preprocess_summaries(reference_summaries, generated_summaries):
    processed_ref_summaries = []
    processed_gen_summaries = []

    for ref_summary in reference_summaries:
        tokenized_ref_summary = [word_tokenize(sentence) for sentence in ref_summary]
        processed_ref_summaries.extend(tokenized_ref_summary)

    for gen_summary in generated_summaries:
        tokenized_gen_summary = word_tokenize(gen_summary)
        processed_gen_summaries.append(tokenized_gen_summary)

    return processed_ref_summaries, processed_gen_summaries


# Preprocess the summaries
processed_ref_summaries, processed_gen_summaries = preprocess_summaries(reference_summaries, generated_summaries)

# Calculate the BLEU scores for each pair of summaries
bleu_scores = []
for ref_summary, gen_summary in zip(processed_ref_summaries, processed_gen_summaries):
    bleu_score = corpus_bleu([[ref_summary]], [gen_summary])
    bleu_scores.append(bleu_score)

# Calculate the average BLEU score
avg_bleu_score = sum(bleu_scores) / len(bleu_scores)
print(f"\nAverage BLEU Score: {avg_bleu_score}")



# Initialize the ROUGE scorer
rouge = Rouge()

# Calculate the ROUGE scores for each pair of summaries
rouge_scores = []
for ref_summary, gen_summary in zip(processed_ref_summaries, processed_gen_summaries):
    ref_summary = ' '.join(ref_summary)
    gen_summary = ' '.join(gen_summary)
    scores = rouge.get_scores(gen_summary, ref_summary)
    rouge_scores.append(scores)

# Calculate the average ROUGE scores
avg_rouge_scores = {}
for metric in rouge_scores[0][0].keys():
    scores = [score[0][metric]['f'] for score in rouge_scores]
    avg_rouge_scores[metric] = sum(scores) / len(scores)

# Print the average ROUGE scores
print("\nAverage ROUGE Scores:")
for metric, score in avg_rouge_scores.items():
    print(f"{metric}: {score}")

```
